Import the packages

```python
import numpy as np
import pandas as pd


# import matplotlib package
import matplotlib as mpl
import matplotlib.pyplot as plt
```

Load the csv file into a DataFrame

```python
df = pd.read_csv('/kaggle/input/canada-per-capita-income/Canada_per_capita_income.csv')
```

```python
# show information of DataFrame df
df.info()
```


create a new figure with size = 10,6 (width, height) in inches using `plt.figure()`

```python
plt.figure(figsize=(10, 6))
```

plot data in line format with marker

```python
X = df["year"]
y = df["income"]
```

```python
plt.plot(X, y, marker='o')
plt.title('Cancada per capita income over years\n', size = 16)
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.grid(True)
plt.xticks(rotation = 45)
plt.show()
```

```python
df.plot('year', 'income', marker = 'o')
plt.title('Cancada per capita income over years\n', size = 16)
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.grid(True)
plt.xticks(rotation = 45)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(X,y,marker = 'o')
ax.set_xlabel("Year")
ax.set_ylabel("per Capita Income")
ax.set_title("Canada per Capita Income over Years")
ax.grid()
plt.xticks(rotation = 45)
plt.show()
```

```python
marketing = pd.read_csv('/kaggle/input/arketing-campaign/marketing_campaign.csv', sep = ';', parse_dates=['Dt_Customer']) 

# Group by multiple columns
Customer_groups = marketing.groupby(['Year_Birth','Education'])['ID'].count()

Customer_groups = pd.DataFrame(Customer_groups.unstack(level=1))
```

```python
Customer_groups.plot()

plt.title('Customer Education Groups\n', size = 16)
plt.xlabel("Year of Birth")
plt.ylabel("Number of Customers")
plt.xlim(1940)
plt.legend(labels = Customer_groups.columns.values)
```
